```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np

estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia)  # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = ['red', 'blue', 'orange', 'green', 'cyan']  # lista en lugar de set, para garantizar
                                                        # que el orden de los colores coincide
                                                        # siempre con el de 'estudios'
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

### Tarea

Implementar con pandas esta misma información (frecuencias absolutas de nivel_estudios, diagrama de barras y diagrama de sectores).

```python
# Tarea: implementar con pandas esta información
import pandas as pd

serie_estudios = pd.Series(nivel_estudios)

# Frecuencias absolutas con pandas
frecuencia_pd = serie_estudios.value_counts()
display(frecuencia_pd)
```

```python
# Diagrama de barras con pandas
frecuencia_pd.plot(kind='bar', ylabel='Frecuencia absoluta', xlabel='Nivel de estudios',
                    title='Diagrama de barras (pandas)', rot=0)
plt.show()
```

```python
# Diagrama de sectores con pandas
frecuencia_pd.plot(kind='pie', autopct='%1.2f%%', ylabel='',
                    title='Diagrama de sectores (pandas)')
plt.show()
```
