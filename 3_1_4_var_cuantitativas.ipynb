```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,
              -28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
print(X)
```

```python
# Medidas de tendencia central
from scipy import stats
media = np.mean(X)
mediana = np.median(X)
moda, _ = stats.mode(X)
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt

Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
plt.boxplot(X)
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True)  # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html

print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

if curtosis > 0:
    print('Distribución Leptocúrtica')
elif curtosis < 0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# Calculamos media y desviación típica de cada variable, agrupadas por clase
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

medias = {}
desvest = {}
for clase in np.unique(y):
    datos_clase = X[y == clase]
    medias[clase] = datos_clase.mean(axis=0)
    desvest[clase] = datos_clase.std(axis=0)

for clase in np.unique(y):
    print(f'--- Clase: {target_names[clase]} ---')
    print('Media    :', np.round(medias[clase], 3))
    print('Desv.Est.:', np.round(desvest[clase], 3))
```

```python
# Gráfico de errorbar: media ± desviación típica de cada variable por especie
plt.figure(figsize=(9, 6))

x_pos = np.arange(len(feature_names))
colores = ['tab:blue', 'tab:orange', 'tab:green']

for clase, color in zip(np.unique(y), colores):
    plt.errorbar(
        x_pos,
        medias[clase],
        yerr=desvest[clase],
        marker='o',
        capsize=4,
        label=target_names[clase],
        color=color
    )

plt.xticks(x_pos, feature_names, rotation=15)
plt.xlabel('Variables')
plt.ylabel('Valor (cm)')
plt.title('Media y desviación estándar de cada variable por especie (Iris)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
